```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data Sets/StudentsPerformance.csv")
df.head()
```

```python
#Line Chart
counts = df['math score'].value_counts().sort_index()

# Define x and y for plotting
x = counts.index
y = counts.values

plt.plot(x, y)
plt.title("Line Chart")
plt.xlabel("Math Score")
plt.ylabel("Count")
plt.show()
```

```python

average_scores = df.groupby('race/ethnicity')['math score'].mean()

# Define x and y for plotting
x = average_scores.index
y = average_scores.values

# Plotting a bar chart
plt.bar(x, y)
plt.title("Bar Chart of Average Math Scores by Ethnicity")
plt.xlabel("Ethnicity")
plt.ylabel("Average Math Score")
plt.show()
```

```python
#Pie Chart

scores_by_ethnicity = df.groupby('race/ethnicity')['math score'].sum()

plt.figure(figsize=(8, 6))
plt.pie(scores_by_ethnicity, labels=scores_by_ethnicity.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Math Scores by Ethnicity')

plt.show()
```

```python
#Histagram

data = df['math score']
plt.hist(data,edgecolor='black')
plt.title('Hitagram')
plt.xlabel('Math Score')
plt.ylabel('Frequancy')
plt.show()
```
