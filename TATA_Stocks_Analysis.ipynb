```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tata = pd.read_csv('/content/drive/MyDrive/Data Sets/TATA Stacks/TATAMOTORS_BSE_01Oct20-01Oct21.csv')
tata.head()
```

```python
tata.describe()
```

```python
tata.isnull().sum()
```

```python
tata.shape
```

```python
counts = tata['Open'].value_counts().sort_index()

# Define x and y for plotting
x = counts.index
y = counts.values

plt.plot(x, y)
plt.title("Line Chart")
plt.xlabel("Open")
plt.ylabel("Count")
plt.show()
```

```python
average = tata.groupby('Date')['Close'].mean()

plt.figure(figsize=(20,6))

# Define x and y for plotting
x = average.index
y = average.values

# Plotting a bar chart
plt.bar(x, y)
plt.title("Bar Chart of Average Close by Date")
plt.xlabel("Date")
plt.ylabel("Close")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 7))

plt.subplot(121)
sns.histplot(data=tata,x='Close',bins=30,kde=True,color='g')

plt.subplot(122)
sns.histplot(data=tata,x='Open',bins=30,kde=True,color='r')

plt.show()
```

```python
sns.histplot(data=tata,x='High',bins=30,kde=True,color='y')
sns.histplot(data=tata,x='Low',bins=30,kde=True,color='b')

plt.show()
```

```python
tata['Time'] = np.arange(len(tata.index))
fig, ax = plt.subplots()
ax.plot('Time', 'Volume', data=tata, color='0.75')
ax = sns.regplot(x='Time', y='Volume', data=tata, ci=None, scatter_kws=dict(color='0.25'))
ax.set_title('Time Plot of Volume');
```
