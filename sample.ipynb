```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```
df = pd.read_csv('/content/drive/MyDrive/Data Sets/Sample.csv')
df.head(10)
```

```
g = df.groupby('City')
g
```

```
city_dfs = {}
for city, group in g:
    city_dfs[city] = group.reset_index(drop=True)
```

```
for city, group in g:
    city_dfs[city] = group.reset_index(drop=True)
```

```


fig, axes = plt.subplots(nrows=len(city_dfs), ncols=1, figsize=(10, 15))

for ax, (city, city_df) in zip(axes, city_dfs.items()):
    city_df.plot(kind='bar', x='Day', y=['Temperature', 'Windspeed'], ax=ax, title=city)
    ax.set_xlabel('Date')
    ax.set_ylabel('Value')
    ax.legend(['Temperature', 'Windspeed'])

plt.tight_layout()
plt.show()
```
